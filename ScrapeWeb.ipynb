#Fetching data from Website using API and transform it using Pandas DataFrame under DATA WRANGLING

Fetch data from json using requests module of python. Owner: @singhsourav0

```python
import requests
import pandas as pd
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&page=1").json()
```

```python

popularity = []
vote_count = []
original_language= []
title = []
vote_average = []
overview = []
release_date =[]

for i in response["results"]:
  popularity.append(i['popularity'])
  vote_count.append(i['vote_count'])
  original_language.append(i['original_language'])
  title.append (i["title"])
  vote_average.append (i["vote_average"])
  overview.append(i["overview"])
  release_date.append(i['release_date'])

d = {'title':title, 'overview':overview, 'release_date':release_date, 'popularity':popularity , 'vote_count':vote_count, 'vote_average':vote_average, 'originl_language':original_language }

df = pd.DataFrame(d)
df
```

```python
final = pd.DataFrame()
for j in range(1 ,30):
   response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&page={}'.format(j)).json()
   popularity = []
   vote_count = []
   original_language= []
   title = []
   vote_average = []
   overview = []
   release_date =[]
   for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append (i["title"])
        vote_average.append (i["vote_average"])
        overview.append(i["overview"])
        release_date.append(i['release_date'])

   d = {'title':title, 'overview':overview, 'release_date':release_date, 'popularity':popularity , 'vote_count':vote_count, 'vote_average':vote_average, 'originl_language':original_language }
   df = pd.DataFrame(d)
   final = pd.concat([final, df], ignore_index=True)
final
```
